```python
import pandas as pd
```

```python
sp_songs = pd.read_csv('song_sp.csv')
en_songs = pd.read_csv('song.csv')
cross = pd.read_csv('cross_ref_no_none.csv')
sp_songs
```

```python
with_cross = en_songs.merge(cross,how='left',left_on='songNumber',right_on='english_number')
with_cross
```

```python
with_sp_names = with_cross.merge(sp_songs,how='left',left_on='spanish_number',right_on='songNumber')
with_sp_names.to_csv('final.csv',index=False)
```
